# 4.0 Data Analysis

```python
#Merging the DataFrames

dfs = pd.concat([Dt_1,Dt_2,Dt_3], ignore_index=True,axis=0)
dfs.head()
```

```python
Telcom_data_geo = cells_geo.merge(dfs,how='inner', left_on=['SITE_CODE'], right_on=['SITE_ID'])
Telcom_data_geo
Telcom_data_geo.drop_duplicates()
```

```python
#Telcom_data_geo_clean = Telcom_data_geo.fillna(0)
Telcom_data_geo_clean = Telcom_data_geo.dropna(how='all')
Telcom_data_geo_clean
Telcom_data_geo_clean.drop_duplicates()
```

```python
#Descriptive analysis summary


#which 3 cities have the highest value

Telcom_data_geo_clean.nlargest(3,['VALUE'])

```

```python
#Cities with the least values

Telcom_data_geo_clean.nsmallest(5,['VALUE'])
```
